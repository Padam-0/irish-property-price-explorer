```python
import pandas as pd
```

```python
df = pd.read_csv("good_data.csv", encoding='latin1', index_col=0)
df.head()
```

```python
df.info()
```

```python
dups = df.address.duplicated()
dup = pd.concat([df.address, df.sale_date, df.price,dups], axis=1).reset_index()
dup.columns = ['uid','address', 'sale_date', 'price', 'bool']
dupl = dup[dup['bool'] == True]
```

```python
len(dupl)
```

```python
dupl = dupl.sort_values(by='address')
dupl
```
